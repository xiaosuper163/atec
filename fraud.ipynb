```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
scaler = StandardScaler()
XTr_scale = scaler.fit_transform(train[features])
XVd_scale = scaler.transform(valid[features])
pca = PCA(n_components=25)
XTr_pca = pca.fit_transform(XTr_scale)
XVd_pca = pca.transform(XVd_scale)
```

```python
from sklearn.ensemble import forest
def set_rf_samples(n):
    """ Changes Scikit learn's random forests to give each tree a random sample of
    n random rows.
    """
    forest._generate_sample_indices = (lambda rs, n_samples:
        forest.check_random_state(rs).randint(0, n_samples, n))
```

```python
def reset_rf_samples():
    """ Undoes the changes produced by set_rf_samples.
    """
    forest._generate_sample_indices = (lambda rs, n_samples:
        forest.check_random_state(rs).randint(0, n_samples, n_samples))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
m = RandomForestClassifier(n_estimators=40, min_samples_leaf=3, max_features=0.5, n_jobs=-1)
m.fit(train[features], train.label)
# print_score(m)
```

```python
m_pca = RandomForestClassifier(n_estimators=40, min_samples_leaf=3, max_features=0.5, n_jobs=-1)
m_pca.fit(XTr_pca, train.label)
# print_score(m)
```

```python
train_pred = m.predict_proba(train[features])[:, 1]
valid_pred = m.predict_proba(valid[features])[:, 1]
```

```python
train_pred_pca = m_pca.predict_proba(XTr_pca)[:, 1]
valid_pred_pca = m_pca.predict_proba(XVd_pca)[:, 1]
```

```python
from sklearn.metrics import roc_auc_score, roc_curve

def my_score2(y_true,y_prob): ##Adapted from SKlearn
    fpr,tpr,threhold = roc_curve(y_true, y_prob) 
    tpr1 = tpr[(fpr>=0.001).argmax()]
    tpr2 = tpr[(fpr>=0.005).argmax()] 
    tpr3 = tpr[(fpr>=0.01).argmax()]
    return 0.4 * tpr1 + 0.3 * tpr2 + 0.3* tpr3

def print_score(train_pred, valid_pred):
    res = [roc_auc_score(train.label, train_pred), roc_auc_score(valid.label, valid_pred), 
           my_score2(train.label, train_pred), my_score2(valid.label, valid_pred)]
    if hasattr(m, 'oob_score_'): res.append(m.oob_score_)
    print(res)
```

```python
print_score(train_pred, valid_pred)
```
